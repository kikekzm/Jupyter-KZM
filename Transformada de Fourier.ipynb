## Detectando la periodicidad.

#### En la Carpeta de datos del Tema 6 del Campus virtual se encuentra el fichero "sunspots.txt" (ya utilizado en el Tema 1), que contiene el número de manchas solares del Sol contabilizadas mes a mes desde Enero de 1749.  El fichero tiene dos columnas, la primera representa el mes y la segunda el número de manchas solares.



```python
import numpy as np
data=np.loadtxt('sunspots.txt')
spots=data[:,1]
time=data[:,0]*2.592e6   #pasamos a segundos
```

###### 1.  Escribir un programa que lea el archivo y genere un gráfico con el número mensual de manchas solares en función del tiempo.   Comprobar que tiene un comportamiento aparentemente cíclico. Estimar la longitud del ciclo en meses.


```python
import matplotlib.pyplot as plt
plt.plot(time,spots)
plt.show()

index1=np.argmax(spots[0:1000])
index2=np.argmax(spots[1000:3000])

T=time[index2]-time[index1]

print('La logitud del ciclo es de aproximadamente',T,'segundos; o ',T/2.592e6,'meses, ',T//(2.592e6*12), 'años.')
```

##### 2. Calcular la transformada de Fourier del número de manchas solares y representar la magnitud  de sus coeficientes al cuadrado $|c_k|^2$.  Es el llamado espectro de potencia de la señal. 

```python
def dft(y):

    N=len(y)
    c=np.zeros(N//2 + 1,complex)
 
    for k in range(N//2+1):
        for n in range(N):
            c[k]+=y[n]*np.exp(-2j*np.pi*k*n/N)
    return c

c=dft(spots)
c_sqr=(c*np.conjugate(c))
print(np.imag(c_sqr))

plt.plot(c_sqr)
plt.xlim(0,200)
plt.show()
```

###### 3. Comprobar que hay un frecuencia dominante y obtener su valor.  ¿Qué es lo que quiere decir? ¿Cual es la periodicidad que se puede extreaer de la señal?

```python
f_max=np.argmax(c)
print('La frecuencia dominante es de',f_max,'Hz.')
print('Su coeficiente asociado es c=',np.real(max(c)))
```
