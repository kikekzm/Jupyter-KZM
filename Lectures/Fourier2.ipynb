## 6.3 Transformadas de cosenos y senos discretas.

- Hasta ahora hemos usado la representación compleja de la transformada de Fourier. 

- Sin embargo sí nuestra función tiene una simetria par o impar podemos usar sólo la representación en cosenos o senos.

- A primera viesta, esto parece restringir de forma evidente su aplicación.

- No obstante, al igual que podemos aplicar la DFT a una función no periodica al repetirla continuamente,   
  podemos hacer algo parecido con las series de senos y cosenos.
  
- Por ejemplo para el serie del coseno, podemos simetrizar una función cualquiera definida en el intervalo $[0,\frac{L}{2}]$ simplemente añadiendo su imágen especular en el intervalo $[\frac{L}{2},L]$.  

- Eso implica que en su versión discreta, el numero de puntos de una muestra siempre va a ser par. 


```python
from IPython.display import Image,display
display(Image("cosine-transform.png"))
```

- Para la serie del seno podríamos hacer exactamente lo mismo. Sin embargo, la antisimetría del seno requeriria que la función se anulase en los extremos del intervalo $[0,\frac{L}{2}]$, lo que hace que sea mucho menos usada.

- Una vez esta función ha sido construida, su transformada de Fourier se puede calcular trivialmente. 

- Además, al ser simetría con respecto a $x=\frac{L}{2}$ tenemos que, dada una partición $x_n=\frac{n L}{N}$ la función satisface que:

$$y_n=f(x_n)=f(x_{N-n})=y_{N-n},$$

$\quad$ que implica que $N$ tiene que ser par. 

- De esta forma, la transformada coseno de la función así definida viene dada por:


\begin{align}
c_k=\sum_{n=0}^{N-1}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}=&\sum_{n=0}^{N/2}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}+\sum_{n=N/2+1}^{N-1}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}\nonumber\\
=&\sum_{n=0}^{N/2}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}+\sum_{n=N/2+1}^{N-1}{y_{N-n}\exp\left(i\frac{2\pi k (N-n)}{N}\right)}\nonumber\\
=&\sum_{n=0}^{N/2}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}+\sum_{n=1}^{N/2-1}{y_{n}\exp\left(i\frac{2\pi k n}{N}\right)}\nonumber\\
=&y_0+y_{N/2}\cos\left(\frac{2\pi k N/2 }{N}\right)+2\sum_{n=1}^{N/2-1}{y_{n}\cos\left(\frac{2\pi k n}{N}\right)}.\nonumber\\
\end{align}

$\quad$ que se denomina **transformada discreta coseno (DCT)** de tipo I (la función es evaluada en los puntos de la muestra).


- Cuando se aplica sobre funciones reales, la mayoría de los casos, los coeficientes $c_k$ son reales, y por tanto cumplen que $c_{N-n}=c_n*=c_n$.

- De esta forma, la transformada inversa coseno de una función real es:


\begin{align}
y_n=\frac{1}{N}\sum_{j=0}^{N-1}{c_k\exp\left(i\frac{2\pi k n}{N}\right)}=&\frac{1}{N}\sum_{k=0}^{N/2}{c_k\exp\left(i\frac{2\pi k n}{N}\right)}+\sum_{k=N/2+1}^{N-1}{c_k\exp\left(i\frac{2\pi k n}{N}\right)}\nonumber\\
=&\frac{1}{N}\sum_{k=0}^{N/2}{c_k\exp\left(i\frac{2\pi k n}{N}\right)}+\sum_{k=N/2+1}^{N-1}{c_{N-k}\exp\left(-i\frac{2\pi (N-k)n}{N}\right)}\nonumber\\
=&\frac{1}{N}\sum_{n=0}^{N/2}{c_k\exp\left(i\frac{2\pi k n}{N}\right)}+\sum_{k=1}^{N/2-1}{c_{k}\exp\left(-i\frac{2\pi k n}{N}\right)}\nonumber\\
=&\frac{1}{N}\left[c_0+c_{N/2}\cos\left(\frac{2\pi n N/2 }{N}\right)+2\sum_{k=1}^{N/2-1}{c_{k}\cos\left(\frac{2\pi k n}{N}\right)}\right],\nonumber\\
\end{align}

$\quad$ que se denomina  **transformada discreta coseno inversa (DCT inversa)** y es equivalente a la transformada directa salvo por el factor $1/N$. 

### 6.3.1 Transformadas de cosenos y senos discretas de tipo II.


- De la misma forma que para las transformadas de Fourier, también es habitual usar las transformadas de senos y cosenos de tipo II, aquellas en las que la función se evalua entre dos puntos de la muestra. 

- Para el caso de la series de cosenos de tipo II, la simetría con respecto al punto medio $x=\frac{L}{2}$ implica que  $y_n=y_{N-1-n}$ por lo que pueden escribirse como:


\begin{align}
c_k=&\sum_{n=0}^{N/2-1}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}+\sum_{n=N/2}^{N-1}{y_n\exp\left(-i\frac{2\pi k n}{N}\right)}\nonumber\\
=&\exp\left(i\frac{\pi k}{N}\right)\left[\,\sum_{n=0}^{N/2-1}{y_n\exp\left(-i\frac{2\pi k(n+1/2)}{N}\right)}+\sum_{n=N/2}^{N-1}{y_{N-n-1}\exp\left(i\frac{2\pi k (N-1/2-n)}{N}\right)}\,\right]\nonumber\\
=&\exp\left(i\frac{\pi k}{N}\right)\left[\,\sum_{n=0}^{N/2-1}{y_n\exp\left(-i\frac{2\pi k (n+1/2)}{N}\right)}+\sum_{n=0}^{N/2-1}{y_{n}\exp\left(i\frac{2\pi k (n+1/2)}{N}\right)}\,\right]\nonumber\\
=&2\exp\left(i\frac{\pi k}{N}\right)\sum_{n=0}^{N/2-1}{y_{n}\cos\left(\frac{2\pi k (n+1/2)}{N}\right)}.\nonumber\\
\end{align}

- En general el factor $\exp\left(i\frac{\pi k}{N}\right)$ se suele reabsorber definiendo un nuevo parámetro

$$a_k=\exp\left(-i\frac{\pi k}{N}\right)c_k=2\sum_{n=0}^{N/2-1}{y_{n}\cos\left(\frac{2\pi k (n+1/2)}{N}\right)},$$

$\quad$ que es la forma habitual de escribir la **transformada de cosenos discreta de tipo II**.

- De la misma forma, la **transformada de cosenos discreta inversa de tipo II** viene dada por:

$$y_n=\frac{1}{N}\left[ a_0+2\sum_{k=1}^{N/2-1}{a_{k}\cos\left(\frac{2\pi k (n+1/2)}{N}\right)}\,\right]$$


- Dada su simplicidad la transformada de cosenos discreta de tipo II es probablemente la versión  más usada.

- Transforma $N/2$ elementos en otros $N/2$. Por eso normalmente se suele sustituir $N/2\to N$ y $a_k\to 2 a_k$ de forma que su expresión se reduce a:


$$a_k=\sum_{n=0}^{N-1}{y_{n}\cos\left(\frac{\pi k (n+1/2)}{N}\right)},\quad y_n=\frac{1}{N}\left[ a_0+\sum_{k=1}^{N-1}{a_{k}\cos\left(\frac{\pi k (n+1/2)}{N}\right)}\,\right],$$

$\quad$ donde ahora, $N$ es el número de puntos de nuestra función inicial, sin simetría par,  que se están transformando.

- La razón por la que se usa más frecuentemente la DCT que la DFT para funciones no periodicas es que mientras que en la DFT, al repetir periodicamente la función, estamos creando discontinuidadades en los extremos del intervalo, la transformada coseno, al simetrizar la función, nos da una función contínua.

- Esas discontinuidades en los extremos de la DFT para funciones no periodicas puede generar problemas, efecto  Gibbs, mientras que tal problema no aparece en la DCT.

- La transformación discreta de seno se obtiene de la misma forma, aunque como ya hemos comentado se aplica de forma menos frecuenta pues requiere que la función se anule en los extremos. 

- Sin embargo, hay problemas físicos en los que es interesnate, por ejemplo para una cuerda vibrante, funciones de onda en un pozo de potencial infinito, etc.


###### 6.3.2 Aplicaciones de la transformada del coseno.

- La DCT tiene usos tecnológicos muy importantes. Por ejemplo es la base del funcionamiento de los archivos comprimidos JPEG, MPEG o MP3.


- Digitalmente una imágen no es más que un conjunto de pixeles (un cuadrado de la pantalla con un color asignado).  
  Un ordenador lo único que necesita es guardar el número que describe (array) el color de cada uno de ellos.   
  Cómo el número de pixeles es muy grande, los archivos que almacenan esa información son muy grandes.
  
  
- El formato JPEG aglutina primero conjuntos de pixeles en bloques y sobre cada uno de ellos aplica una DCT de tipo II.   
  Con ello obtiene todos los coeficientes $a_k$ asociados a la imagen, pero sólo almacena los más grandes.    
  Esto permite dismiuir el tamaño del fichero de almacenamiento. 


- El formato MPEG hace lo mismo, pero sobre cada una de las imágenes que genera un video. 


- El formato de audio MP3 funciona de forma parecida. Digitalmente una señal de sonido no es más que un conjunto de ondas espaciadas en el tiempo.  
  El formato MP3 aplica a cada una de esas muestras una DCT y sólo guarda aquellas frecuencias que son perceptibles para el hombre.



## 6.4 La transformada de Fourier rápida.

- La transformada de Fourier discreta de una función real

$$c_k=\sum_{n=0}^{N-1}y_n\exp{\left(-\frac{2\pi k n}{N}\right)},$$

$\quad$ involucra $\#N\cdot(\frac{1}{2}N+1)\sim\frac{1}{2}N^2$ operaciones.

- En el Tema 2 ya vimos que el número de operaciones que se pueden hacer en un ordenador es de aproximadamente $10^9$.

- Eso implica que el tamaño máximo de una muestra a la que se le puede aplicar la DFT es de:

```python
from math import sqrt

n_oper=1e9
N=sqrt(2*n_oper)
print("%0.f" %N)
```

- Este es aproximadamente el número de puntos que se necesita para representar un segundo de una señal musical.


- Por tanto, a la hora de hacer práctico el análisis de Fourier es necesario encontrar otro algorimto para calcular sus coeficientes.


- El algoritmo usado fue desarrollado C. J. Gauss en 1805 y redescubierto por Cooley y Turkey en 1965.   
  Es conocido como la transformada de Fourier rápida (FFT).
  

- Para explicarlo:

    1. Asumamos que $N=2^m$ (de forma general no es necesario este requisito pero simplifica su descripción).
    2. Dividamos nuestro muestra inicial de $N$ puntos en 2 submuentras con valores alternos de la función,  
       es decir, una con los valores pares y otra con los impares de la muestra.  
       Para la submuestra par tenemos:
       
$$c^\text{par}_k=\sum_{r=0}^{N/2-1}{y_{2r}\exp\left(-i\frac{2\pi k (2r)}{N}\right)}=\sum_{r=0}^{N/2-1}{y_{2r}\exp\left(-i\frac{2\pi k r}{N/2}\right)},$$

$\quad\quad\;\;$  mientras que para la impar el resultado es:
       
$$\hat c^\text{impar}_k=\sum_{r=0}^{N/2-1}{y_{2r+1}\exp\left(-i\frac{2\pi k (2r+1)}{N}\right)}=\exp\left(-i\frac{2\pi k}{N}\right)\sum_{r=0}^{N/2-1}{y_{2r+1}\exp\left(-i\frac{2\pi k r}{N/2}\right)}=\text{e}^{\left(-i\,2\pi k/N\right)}c^\text{impar}_k.$$


$\quad\quad\;\;$ Cada una de ellas no es más que la transformada de Fourier de una muestra con $N/2$ puntos, y la transformada de Fourier completa es
  
$$c_k=c^\text{par}_k+\text{e}^{\left(-i\,2\pi k/N\right)}c^\text{impar}_k.$$

$\quad$ 3. Repetimos iterativa el proceso para cada uno de nuestras submuestras, dividiendolas en otras dos.   
$\qquad\!$ De forma que después de $k$ pasos tendremos $2^k$ submuestras con $N/k$ elementos cada una.  
$\qquad\!$ Al llegar al paso $m$, con $m$ tal que $N=2^m$, tendremos $N$ submuestras de un elemento, cuya DFT es:

$$c_0=\sum_{n=0}^0y_n\text{e}^0=y_0,$$

$\qquad\!$ es decir, coincide con los elementos de la muestra. 

- La transformada de Fourier rápida se calcula recostruyendo la DFT a través de los elementos de la muestra.   
  Es decir, aplicando el proceso inverso al anterior. 

    1. Dados los elementos de la muestra, estos se combinan en pares siguiendo la ecuación de arriba. 
    2. Estas combinaciones de pares se combinan en conjuntos sucesivos de 4 elementos.   
       Estos en conjuntos de 8 elementos y así sucesivamente hasta que hemos reconstruido el conjunto total.
    3. Lo relevante es calcular el número de operaciones que hemos necesitado para recuperar la DFT completa.  
       En primer lugar tenemos $N$ muestras de 1 elemento.         
       A continuación, tenemos $N/2$ muestras cada una con 2 coeficientes, lo que requiere otra vez $N$ operaciones. 
       En la tercera ronda, tenemos $N/4$ muestras formadas con 4 elementos cada una, que implica de nuevo $N$ coeficientes. 
       Finalmente en la ronda final $m$, tenemos $N/2^m=1$ muestra con $2^m$ elementos.  
       Por tanto, en cada ronda siempre tenemos que realizar $N$ operaciones.  
       Falta calcular cuantos rodas necesitamos:
       
$$m=\log_2 N,$$

- Que por tanto implica que el número total de operaciones que hemos necesitado para recuperar la DFT es de:
       
$$\#(\text{operaciones})= N\cdot m=N\cdot\log_2N.$$

- Para $10^9$ operaciones el tamaño de la muestra es:

```python
from math import log
from scipy.optimize import fsolve

N_oper=[1e9]

def f(x):
    return x*log(x,2)-N_oper


print("%0.1e" %fsolve(f,1e6))
```

- Que implica que ahora podemos trabajar con muestras de hasta 40 millones de elementos.

### 6.4.1  Formulas para la transformada de Fourier rápida.

- Como ya hemos visto, en el paso inicial tenemos la DFT completa que queremos recuperar, una muestra con $N$ elementos. 
- Para el paso siguiente, tendremos 2 muestras con $N/2$ elementos cada una.
- Para un paso $n$ tendremos $2^n$ conjuntos con $N/2^n$ elementos cada uno.
<br/>

    - El primero de estos conjuntos constará de los elementos de la muestra inicial: $\{0,2^n, 2\cdot 2^n,3\cdot 2^n,\cdots,N/2^n-1\}$.  
      Es decir por los elementos $2^n\cdot r$, con $r=0,\cdot,N/2^n-1$
    - De la misma forma, la segunda muestra contará con los elementos: $2^n\cdot r+1$, la tercera con $2^n\cdot r+2$, etc... 
    - Por tanto, la DFT de la muestra $j$ del paso $n$ será:
    
    
$$\sum_{r=0}^{N/2^n-1}y_{2^n r+j}\exp{\left(-i\frac{2\pi k(2^n r +j)}{N}\right)}=\text{e}^{-i2\pi kj\,/N}\sum_{r=0}^{N/2^n-1}y_{2^n r+j}\exp{\left(-i\frac{2\pi k r)}{N/2^n}\right)}=\text{e}^{-i2\pi kj\,/N}c_k^{\,(n,j)}.$$

$\qquad\;\;\;$ con $k,j=0,\cdots,N/2^n-1.$ 
 
- Pero ya hemos visto que esa DFT se puede descomponer en 2 submuestras con la mitad de elementos de forma que:
   
$$c_k^{(n,j)}=c_k^{(n+1,\,j)}+\text{e}^{-i2\pi k 2^n\,/N}c_k^{(n+1,\,j+2^n)},$$

- Con esta formula el cálculo de la transformada de Fourier rápida (FFT) es fácil:


- Empezamos con el último paso $m$ con $N=2^m$, en ese caso, sólo hay un coeficiente de Fourier para nuestras $N$ muestras
   
$$c_0^{(m,j)}=y_j, \quad\text{con}\quad j=0,\cdots, N-1,$$

-  Usamos estos valores para construir los coeficientes de Fourier de la fase anterior, hasta llegar a $m=0$ en donde:
    
$$c_{k}^{0,0}=\sum_{r=0}^{N-1}y_n\exp{\left(-i\frac{2\pi k r}{N}\right)}=c_k,$$   

$\quad$ tenemos los coeficientes que estabamos buscando.

### 6.4.2 La transformada de Fourier rápida en numpy.

- Cómo no podía ser de otra forma ya hay un módulo en numpy en el que la transformada de Fourier rápida esta incorporada. Su implementación es además mucho más eficiente de la que hemos programado nosotros. 

- Es el módulo fft. 

- Este módulo incorporá las funciones rfft y irfft para la transformada de Fourier rápida de una muestra real y su inversa, respectivamente.

- También incorpora las funciones fft y ifft, en este caso para una muestra compleja. 

- Para muestras de dos dimensiones incorporá las funciones rfft2, irfft2 y fft2, ifft2. 

- Para funciones de más de dos dimensiones la función se llama rfftn, irfftn, y fftn, ifftn.

```python
from numpy import array, arange
from numpy.fft import rfft

y = arange(0,1,0.1)
c = rfft(y)
print(c)
```

### 6.4.3 Transformada Coseno y Seno Rápidas.

- La transformada del coseno y seno se puede obtener trivialmente usando la FFT pues no son  mas que una transformada de Fourier con una muestra simétrica o antisimétrica, respectivamente.


- Método para calcular una FCT y una FST:

   - Dada una muestra de N puntos, creamos una muestra con el doble de puntos por simetría especular par o impar dependiendo de si queremos una transformada seno o coseno.
   
   - Se realiza una FFT sobre la muestra obtenida.

**Ejercicio 6.5:** Escribir un programa que calcula la FCT y la FST directa e inversa de tipo II. 

```python
from numpy.fft import rfft
from numpy import empty,linspace,exp,real,pi

def dct(y):
    
    N = len(y)                 # Tamaño de la muestra
    y2 = empty(2*N,float)      # Creamos una nueva array con dimensión doble 
                               # para guardar la señal simetrizada    
    for n in range (N):
        y2[n] = y[n]           # la construimos
        y2[2*N-1-n] = y[n]
        c = rfft(y2)
        phi = exp(-1j*pi*arange(N)/(2*N))  # factor de fase que nos convierte la señal coseno de tipo II
    
    return real(phi*c[:N])


def idct(a):                          # hacemos lo mismo pero con la transformación inversa
    
    N = len(a)
    c = empty(N,complex)
    
    phi = exp(1j*pi*arange(N)/(2*N))
    c[:N] = phi*a
    return irfft(c)[:N]
```

```python
from numpy.fft import rfft
from numpy import empty,linspace,exp,imag,pi

def dst(y):
    
    N = len(y)                 # Tamaño de la muestra
    y2 = empty(2*N,float)      # Creamos una nueva array con dimensión doble 
                               # para guardar la señal simetrizada   
            
    y2[0]=y2[N]=0              # para hacer la transformada seno es necesario imponer que
                               # la función se anule en los extremos                
        
    for n in range (N):
        y2[n] = y[n]           # la construimos
        y2[2*N-1-n] = -y[n]                                    
        
        c = rfft(y2)
        phi = exp(-1j*pi*arange(N)/(2*N))  # factor de fase que nos convierte la señal coseno de tipo II

    a=-imag(phi*c[:N])
    a[0]=0
    return a

def idst(a):
    
    N=len(a)
    c=empty(N+1,complex)
    c[0]=c[N]=0
    c[1:N]=-1j*a[:N]
    y=irfft(c)[:N]
    y[0]=0
    
    return y
```

**Ejemplo 6.7: Deconvolución de imagenes.** 

Estamos habituados a ver en películas o series de televisión de temática policiaca escenas en las que un técnico transforma una imágen borrosa en una imagen nítida casi por arte de magia. 
Pero de hecho, no hay ninguna magia, no es más que analisis de Fourier.  

En una imagen borrosa cada punto queda difuminado por medio de una distribucioń que llamamos función de dispersión de punto.  
Para simplificar el problema asumamos que estamos trabajando con una imagen en blanco y negro, por lo que la imagen puede ser representada por una sóla función $a(x,y)$ que determina el brillo de un punto $(x,y)$ de la imagen.

Definamos la función de dispersión de punto como $f(x,y)$, que hace que dado un punto $(x,y)$ en la imagen inicial, su brillo (valor) se traslade al punto $f(x,y)$. Sí $f(x,y)$ es una función ancha, la imagen se emborronara mucho. Sin embargo, si la función esta descrita por un pico muy estrecho entonces la imagen puede ser aceptablemente nítida.

En general, el brillo de la imagen borrosa $b(x,y)$ puede describirse por una función de convolución, que describe como se ha desplazado el  brillo de la imagen original:

$$b(x,y)=\int_{0}^K\int_0^L a(x',y')f(x-x',y-y')\,\text{d}x'\text{d}y',$$

donde $K$ y $L$ es la dimensión de la imagen en cada uno de sus ejes. Esta ecuación es la llamada convolución de la imagen a través de la función de dispersión de punto.

Para explicar como funciona la técnica de deconvolución (recuperación de una imagen borrosa) trabajemos primero en una dimensión (la generalización a dimeniones es trivial). La convolución de una imagen en una dimension viene dada por:

$$b(x)=\int_0^L a(x')f(x-x')\,\text{d}x'.$$

Esta función se puede representar por medio de una serie de Fourier de forma que:

$$b(x)=\sum_{k=-\infty}^\infty{\bar b_k\exp\left(i\frac{2\pi k x}{L}\right)},$$

donde los coeficientes de Fourier vendrán dado por:

\begin{align}
\bar b_k=&\frac{1}{L}\int_0^L{b(x)\exp{\left(-i\frac{2\pi k x}{L}\right)}\text{d}x}\nonumber\\
=&\frac{1}{L}\int_0^L\int_0^La(x')f(x-x')\,\exp{\left(-i\frac{2\pi k x}{L}\right)}\,\text{d}x\,\text{d}x'\nonumber\\
=& \frac{1}{L}\int_0^La(x')\,\exp{\left(-i\frac{2\pi k x'}{L}\right)}\int_{-x'}^{L-x'}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x\,\text{d}x'',
\end{align}

donde en el último paso hemos llamado $x''=x-x'$.

Si ahora hacemos $f(x)$ una función periodica repitiendola infinitamente hacía la izquierda y derecha de su intevalo, la segunda integral

\begin{align}
\int_{-x'}^{L-x'}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''=&\int_{-x'}^{0}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''+\int_{0}^{L-x'}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''\nonumber\\
=&\int_{L-x'}^{L}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''+\int_{0}^{L-x'}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''\nonumber\\
=&\int_{0}^{L}f(x'')\,\exp{\left(-i\frac{2\pi k x''}{L}\right)}\,\text{d}x''=L \bar f_k,
\end{align}

no es más que la transformada de Fourier $\bar f_k$ de $f(x)$ multiplicada por $L$.

Por tanto, 
$$\bar b_k=\bar f_k\int_0^La(x')\,\exp{\left(-i\frac{2\pi k x'}{L}\right)\,\text{d}x}=L\bar f_k\bar a_k,$$

la transformada de Fourier de la imagen borrosa viene dada por el producto de la transformada de fourier de la imagen original y de la función de dispersión de punto.

Para recuperar la imagen original por tanto sólo tenemos que coger la imagen borrosa y hacer su transformada de Fourier. Hacer lo mismo con la función de dispersión de punto y tomar su cociente dividido por el tamaño del intervalo, de forma que obtenemos

$$\bar a_k=\frac{\bar b_k}{L \bar f_k}.$$

Finalmente sólo tenemos que hacer la transformada de Fourier inversa para recuperar la imagen nítida. 
Este es llamado proceso de deconvolución.

El proceso para una imagen de dos dimensiones es completamente equivalente. Para una imagen de dimensiones $L\times K$ se obtiene

$$\bar b_{kl}=KL\bar a_{kl}\bar f_{kl}.$$


La única complicación asocidada al proceso de deconvolución de una imagen es que a priori no sabemos cual es la función de dispersión del punto, por lo que uno tiene que probar con varias alternativas hasta que obtiene un resultado satisfactorio.

La mayoría de las cámaras, sin embargo, involucran una función de dispersión del punto que puede ser descrita de forma bastante satisfactoria como una función gaussiana:

$$f(x,y)=\exp\left(-\frac{x^2+y^2}{2\sigma^2}\right),$$

donde el valor de la desviación típica $\sigma$ es desconocido.
Este es por ejemplo el procedimiento que se uso para obtener la primera imagen nítida de un agujero negro.
