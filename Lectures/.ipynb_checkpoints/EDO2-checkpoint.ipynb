## 7.1 Ecuaciones diferenciales de primer orden en una variable.

### 7.1.2 Métodos de Runge-Kutta.


- Dada la incetidumbre del método de Euler, uno se puede plantear varias estrategías para aumentar su precisión.


- La primera de ellas es incluir más términos en el desarrollo de Taylor de la solución, lo cual requiere evaluar la derivada de la función. Esto es problemático porque:

    1. En muchas ocaciones no conocemos la derivada de la función. 
    2. Incluso si lo hacemos, trabajar con ella vuelve el método menos eficiente y en muchos casos menos preciso.
   
   
- La segunda alternativa esta asociada a los métodos que vimos para aumentar la precisión de las derivadas numéricas:  
  escoger los puntos en los que evaluamos la solución para que al combinarlos, los términos superiores del desarrollo de Taylor se cancelen.
  
  
- Esta segunda alternativa es la base de los métodos de Runge-Kutta. 

####  Método de Runge-Kutta de segundo orden o método del punto medio.

- A primer oden, el método de Runge-Kutta es el método de Euler. La derivada de la solución en la serie de Taylor en el punto $t+h$ se evaluar en $t$.

- A segundo orden, el método de Runge-kutta evalua la función en el punto que garantiza que el término ${\cal O}(h^2)$ de la serie de Taylor se cancela.

- Es algo que se puede representar mediante la figura:  

```python
from IPython.display import display,Image
display(Image("Runge-kutta.png",width=600))
```

- Mientras que en el método de Euler la derivada de la solución se evalua en $t$, introduciendo un error proporcional a su segunda derivada,   
  este error ${\cal O}(h^2)$ se puede cancelar si se evalua en el punto medio.
  
- Para verlo repitamos el desarrollo de Taylor en el instante $t+h$ en torno a $t+h/2$:


$$x\left(t+h\right)=x\left(t+\frac{1}{2}h\right)+\frac{1}{2}h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+\frac{1}{8}h^2\left(\frac{\text{d}^2x}{\text{d}t^2}\right)_{t+h/2} +{\cal O}\left(h^3\right),$$

$\quad$ aplicandoselo de nuevo a la condición incial:

$$x\left(t\right)=x\left(t+\frac{1}{2}h\right)-\frac{1}{2}h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+\frac{1}{8}h^2\left(\frac{\text{d}^2x}{\text{d}t^2}\right)_{t+h/2} +{\cal O}\left(h^3\right),$$

$\quad$ por lo que al tomar su diferencia obtenemos:

$$x\left(t+h\right)=x\left(t\right)+h\left(\frac{\text{d}x}{\text{d}t}\right)_{t+h/2}+{\cal O}\left(h^3\right)=x\left(t\right)+hf\left(x\left(t+h/2\right),t+h/2\right)+{\cal O}\left(h^3\right),$$

$\quad$ que proporciona una estimación de la solución en el tiempo $t+h$ con un error $\cal O\left(h^3\right)$.

- Sin embargo, esta ecuación requiere evaluar la función en $x\left(t+h/2\right)$, un valor que disponemos. Para estimarla usamos el método de Euler:

$$x\left(t+\frac{1}{2}h\right) = x(t)+\frac{1}{2}h\,f(x,t)+ \cal O\left(h^2\right),$$


- De esta forma, el conjunto de ecuaciones necesarias para evaluar el método son:

\begin{align}
k_1=&h\,f\left(x,t\right),\nonumber\\
k_2=&h\,f\left(x+\frac{1}{2}k_1,t+\frac{1}{2}h\right),\nonumber\\
x(t+h)=&x(t)+k_2.
\end{align}

$\quad$ que aplicadas secuencialmente e iterativamente a partir de una condición incial, nos permiten obtener nuestra soluciónen todo el intervalo de interés. 

- Sin embargo para obtener nuestro algoritmo hemos aproximado la solución en $t+h/2$ con el método de Euler.   
  Para ver el efecto de tal aproximación, expandamos en serie de Taylor la función en el punto $x'(t)=x(t)+k_1/2$ en torno a a $x(t+h/2)$ en un tiempo $t+h/2$:

$$f\left(x'(t),t+h/2\right)=f\left(x\left(t+h/2\right),t+h/2\right)+\left[x'(t)-x\left(t+h/2\right)\right]\left(\frac{\partial f}{\partial x}\right)_{\left(x\left(t+h/2\right),t+h/2\right)}+{\cal O}\left(\left[x'(t)-x\left(t+h/2\right)\right]^2\right).$$

- Por otro lado, tenemos que:

$$x\left(t+h/2\right)=x(t)+h/2 f(x,t)+{\cal O}(h^2)=x(t)+k_1/2+{\cal O}(h^2)=x'(t)+{\cal O}(h^2),$$

$\quad$ que implica por tanto que 

$$x'(t)-x\left(t+h/2\right)=\cal O(h^2)$$

$\quad$ y 


$$f\left(x'(t),t+h/2\right)=f\left(x\left(t+h/2\right),t+h/2\right)+\cal O(h^2)$$

$\quad$ que una vez introducida en la sólución para $x(t+h)$ proporciona un nuevo témrino ${\cal O} (h^3)$. 


- El método de Runge-Kutta de segundo orden es para cada paso una aproximación de segundo orden, correcto hasta orden $h^2$ con error $\cal O (h^3)$.

- Por el contrario el método de Euler es de primer orden, correcto para cada paso hasta orden $h$ con error $\cal O (h^2)$.

####  Método de Runge-Kutta de oden $n$.

- Veamos como se puede obtener de forma general las ecuaciones del método Runge-Kutta a un órden abitrario.

- Una vez más, dada la ecuación diferencial

$$x'=\frac{\text{d}x}{\text{d}t}=f(x,t),$$

$\quad$ expandimos la solución en serie de Taylor


$$x(t+h)=x(t)+h\,x'+\frac{1}{2}\,h^2\,x''+\frac{1}{3!}\,h^2\,x^{(3)}+\frac{1}{4!}\,h^4\,x^{(4)}+\cdots,$$

$\quad$ con

$$x^{(n)}(t)=\frac{\text{d}^nx(t)}{\text{d}t^n}.$$

- Por otro lado, denotando como:

$$f^{(n,m)}(x,t)=\frac{\partial^{n+m}f(x,t)}{\partial x^n\,\partial t^m},$$

$\quad$ tenemos que cada término en el desarrollo de Taylor de nuestra solución puede escribirse como:

\begin{align}
x'(t)=&f(x,t),\nonumber\\
x''(t)=&f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t),\nonumber\\
x^{(3)}(t)=&f^{(0,2)}(x,t)+2 f^{(1,2)}(x,t)f(x,t)+f^{(0,2)}(x,t)f(x,t)^2+f^{(0,1)}(x,t)^2f(x,t),\nonumber\\
\vdots\quad=&\qquad\qquad\vdots\qquad\qquad\qquad\;\;\;\,\vdots\qquad\qquad\qquad\quad\vdots\nonumber\\
\end{align}


- De la misma forma, la solución en un punto $t+h$ también puede obtenerse combinando la evaluación de la función en varios puntos:

$$x(t+h)=x(t)+\alpha_1\,k_1+\alpha_2\,k_2+\cdots+\alpha_m\,k_m,$$

$\quad$ donde los valores de $k_i$ vienen dados por:

\begin{align}
k_1=&hf\left(x,t\right),\nonumber\\
k_2=&hf\left(x+\nu_{21}k_1,t+\nu_{21}h\right),\nonumber\\
k_3=&hf\left(x+\nu_{31}k_1+\nu_{32}k_2,t+\nu_{21}h+\nu_{22}h\right),\nonumber\\
\vdots=&\qquad\vdots\qquad\qquad\qquad\vdots\nonumber\\
k_m=&hf\left(x+\sum_{i=1}^{m-1}\nu_{mi}k_i,\sum_{i=1}^{m-1}t+\nu_{mi}h\right),\nonumber\\
\end{align}

$\quad$ con $\alpha_i$ y $\nu_{ij}$ parámetros que es necesario determinar.

- Con ello ya tenemos todos los elementos para obtener nuestra aproximación de Runge-Kutta a un orden dado.

- Por ejemplo, para recuperar el resultado de orden 2, tenemos por un lado:

$$x(t+h)=x(t)+h f(x,t)+\frac{1}{2}h^2\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right).$$

$\quad$ y por otro

$$x(t+h)=x(t)+\alpha_1 k_1+\alpha_2 k_2 =x(t)+\alpha_1 h f(x,t)+hf\left(x+\nu_{21}k_1,t+\nu_{21}h\right).$$

- Desarrollando ahora $k_2$, tenemos:

$$k_2=hf(x,t)+h^2 \nu_{21}\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right),$$

$\quad$ que nos permite escribir:


$$x(t+h)=x(t)+\left(\alpha_1+\alpha_2\right)hf(x,t)+h^2 \alpha_2\nu_{21}\left(f^{(0,1)}(x,t)+f^{(1,0)}(x,t)f(x,t)\right).$$

- Comparando ambas expresiones obtenemos que:

$$\alpha_1+\alpha_2=1,\quad \alpha_2\nu_{21}=\frac{1}{2},$$

$\quad$ que es un sistema de dos ecuaciones y tres incognitas, lo cual nos permite fijar una libremente.

- Recapitulando, tenemos el sistema de ecuaciones

\begin{align}
\alpha_1+\alpha_2=&\,1,\nonumber\\
\alpha_2\nu_{21}=&\,\frac{1}{2},\nonumber\\
k_1=&\,hf(x,t),\nonumber\\
k_2=&h\,f\left(x+\nu_{21}k_1,t+\nu_{21}h\right),\nonumber\\
x(t+h)=&\,x(t)+\alpha_1 k_1+\alpha_2 k_2,\nonumber\\
\end{align}

$\quad$ que nos proporciona una formula iterativa para encontrar nuestra solución en cualquier punto.

- Por ejemplo, haciendo $\alpha_1=0$, tenemos que $\alpha_2=1$ y $\nu_{21}=1/2$, y nuestra solución es:

\begin{align}
k_1=&\,hf(x,t),\nonumber\\
k_2=&h\,f\left(x+k_1/2,t+h/2\right),\nonumber\\
x(t+h)=&\,x(t)+ k_2,\nonumber\\
\end{align}

$\quad$ que es el método del punto medio que hemos visto.

- Sin embargo hay otras opciones. Cogiendo $\alpha_1=1/2$, tenemos que $\alpha_2=1/2$ y $\nu_{21}=1$ que propociona:

\begin{align}
k_1=&\,hf(x,t),\nonumber\\
k_2=&h\,f\left(x+k_1,t+h\right),\nonumber\\
x(t+h)=&\,x(t)+\frac{1}{2} k_1+\frac{1}{2}k_2,\nonumber\\
\end{align}

$\quad$ que es otro método de Runge-Kutta de orden 2 llamado método de Euler modificado.

####  Método de Runge-Kutta de cuarto orden.

- Los algortimos para los métodos de Runge-Kutta se vuelven más y más complicados a medida que aumentamos el orden.

- En general se considera que el de orden 2 es el que optimiza el cociente preción versus complejidad.

- En particular, el algoritmo de cuarto orden más usado viene dado por el sistema de ecuaciones:

\begin{align}
k_1=&\,hf(x,t),\nonumber\\
k_2=&h\,f\left(x+k_1/2,t+h/2\right),\nonumber\\
k_3=&h\,f\left(x+k_2/2,t+h/2\right),\nonumber\\
k_4=&h\,f\left(x+k_3,t+h\right),\nonumber\\
x(t+h)=&\,x(t)+\frac{1}{6}\left(k_1+2k_2+2k_3+k_4\right).\nonumber\\
\end{align}

$\quad$ que es preciso hasta el orden $h^4$, es decir su error es ${\cal O}(h^5)$.

- El método de Runge-Kutta es el más usado para resolver ecuaciones diferenciales ordinarias, y en general el más fiable para problemas generales.

### 7.1.2 Soluciones sobre rangos infinitos.

- Hasta ahora hemos buscado la solución en un determinado intervalo finito, pero en ocasiones queremos encontrar una solución hasta $t=\infty$.

- Al igual que para la integrales, la forma de hacerlo es usar un cambio de variable. 

- Por ejemplo, haciendo:

$$u=\frac{t}{1+t}\Rightarrow t=\frac{u}{1-u},$$

$\quad$ que cumple que para $t\to\infty$ la nueva variables vale $u\to1$, nuestra ecuación diferencial se puede escribir como

$$\frac{\text{d}x}{\text{d}u}\frac{\text{d}u}{\text{d}t}=f(x,t)\;\Rightarrow\; \frac{\text{d}x}{\text{d}u}=\frac{\text{d}t}{\text{d}u}f\left(x,\frac{u}{(1-u)}\right)=\left(1-u\right)^{-2}f\left(x,\frac{u}{(1-u)}\right).$$

- Por tanto, llamando:

$$g(x,u)=\left(1-u\right)^{-2}f\left(x,\frac{u}{(1-u)}\right),$$

$\quad$ nuestro problema inicial se puede escribir como:

$$\frac{\text{d}x}{\text{d}u}=g(x,u),$$

$\quad$ que es una nueva ecuación de primer oden. Al resolverla en el intervalo $u\in[0,1]$ estamos obteniendo nuestra ecuación orginal para $t\in[0,\infty)$.

**Ejercicio 7.5: Ecuaciones diferenciales en rangos infinitos.**

Dada la ecuación diferencial

$$\frac{\text{d}x}{\text{d}t}=\frac{1}{x^2+t^2},$$

usar el método de  Runge-Kutta de cuarto orden para obtener su solución en el intevalo $t\in[0,\infty)$ con la condición inicial x(0)=1.  
Representarla en el intervalo [0,100].



Haciendo el cambio de variable $t=u\,/\,(1-u)$ tenemos:

$$\frac{\text{d}x}{\text{d}u}=\frac{1}{\left(1-u\right)^2x^2+u^2}=g(x,u),$$

con $u\in[0,1]$.

```python
from numpy import linspace,arange
from matplotlib.pyplot import plot,xlabel,ylabel,xlim,show

# definimos nuestra función en función de u

def g(x,u):
    return 1/(x**2*(1-u)**2+u**2)

a=0.0       # límite inferior en u
b=0.99999   # límite superior en u
N=100       # número de puntos 
h=(b-a)/N   # paso

up=linspace(a,b,N)
tp=[]
xp=[]

x=1
for u in up:
    tp.append(u/(1-u))
    xp.append(x)
    k1=h*g(x,u)
    k2=h*g(x+k1/2,u+h/2)
    k3=h*g(x+k2/2,u+h/2)
    k4=h*g(x+k3,u+h)
    x+=(k1+2*k2+2*k3+k4)/6
    
plot(tp,xp)
xlim(1,100)
xlabel("t")
ylabel("x(t)")
show()
```

## 7.2  Sistemas de ecuaciones diferenciales ordinarias.

- Estudiemos ahora el caso en el que tenemos más de una variable (variable dependientes) dependiendo de un mismo parámetro (variable independiente), que en general asumiremos juega el papel del tiempo.

- Es decir, tenemos un sistema de ecuaciones diferenciales con dos variables $x(t)$, $y(t)$ que són exclusivamente función de $t$:

$$\frac{\text{d}x}{\text{d}t}=f_x(x,y,t),\quad \frac{\text{d}y}{\text{d}t}=f_y(x,y,t),$$

$\quad$ siendo $f_x$ y $f_y$ dos funciones, linéales o no linéales, de las variables $x,y,t$.

- En general, para un número arbitrario de variables y ecuaciones, el sistema se puede escribir en forma vectorial como:

$$\frac{\text{d}\bf{r}}{\text{d}t}={\mathbf F}({\bf r},t),\quad\text{con}\quad{\bf r}=\left(x,y,\cdots\right),\quad\text{y}\quad{\mathbf F}({\bf r},t)=\left(f_x({\bf r},t),f_x({\bf r},t),\cdots\right),$$

$\quad$ que nos permite interpretarla como una función de una sóla variable y analizar su solución como hemos hecho hasta ahora. 

- Por ejemplo, desarrollando la solución en ${\bf r}(t+h)$ serie de Taylor tenemos:

$${\bf r}(t+h)={\bf r}(t)+h\frac{{\text d}{\bf r}(t)}{\text{d}t}+{\cal O}\left(h^2\right)={\bf r}(t)+h\,{\mathbf F}({\bf r},t)+{\cal O}\left(h^2\right),$$

$\quad$ que despreciando el término en $h^2$ nos lleva a la generalicación del método de Euler para más de una variable. 

- De la misma forma, el método de Runge-Kutta de cuarto orden para sistemas de ecuaciones diferenciales vendrá dado por:

\begin{align}
{\bf k_1}=&h\,{\bf F}({\bf r},t),\nonumber\\
{\bf k_2}=&h\,{\bf F}({\bf r}+{\bf k_1}/2,t+h/2),\nonumber\\
{\bf k_3}=&h\,{\bf F}({\bf r}+{\bf k_2}/2,t+h/2),\nonumber\\
{\bf k_4}=&h\,{\bf F}({\bf r}+{\bf k_3},t+h),\nonumber\\
{\bf r}\left(t+h\right)=&r(t)+\,\frac{1}{6}({\bf k_1}+2{\bf k_2}+2{\bf k_3}+{\bf k_4}),\nonumber\\
\end{align}

- Estas ecuaciones pueden implementarse directamente en python haciendo uso de que es un lenguaje vectorial: permite operar con vectores, que pueden jugar el paper tanto de argumentos como resultado de una función.

**Ejercicio 7.6: Sistemas de ecuaciones diferenciales ordinarias.**

Dado el sistema de ecuaciones diferenciales

$$\frac{\text{d}x}{\text{d}t}=xy-x,\quad \frac{\text{d}y}{\text{d}t}=y-xy+\sin^2\omega t,$$

con $\omega=1$, encontrar su solución en el intervalo $t\in[0,10]$, con x(0)=y(0)=1.

```python
from numpy import sin,array,linspace
from matplotlib.pyplot import plot,xlabel,show

# definimos nuestra función como un array, que tiene como argumentos r, otro array, y el tiempo.

def f(r,t):
    x=r[0]
    y=r[1]
    fx=x*y-x
    fy=y-x*y+sin(t)**2
    
    return array([fx,fy],float)


a=0            # punto inicial del intervalo
b=10           # punto final del intervalo
N=1000         # número de puntos
h=(b-a)/N      # tamaño del paso

tp=linspace(a,b,N)     # inicializamos nuestras soluciones para un array de puntos en t
xp=[]
yp=[]

r=array([1,1],float)   # condiciones iniciales para nuestras soluciones

for t in tp:
    xp.append(r[0])
    yp.append(r[1])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)

    r += (k1+2*k2+2*k3+k4)/6

plot(tp,xp)
plot(tp,yp)
xlabel("t")
show()

```

**Ejercicio 7.8: el atractor de Lorenz**

El atractor de Lorenz es un sistema dinámico determinista tridimensional no lineal dado por la ecuaciones:

$$\frac{\text{d}x}{\text{d}t}=\sigma(y-x),\quad \frac{\text{d}y}{\text{d}t}=rx-y-xz,\quad \frac{\text{d}z}{\text{d}t}=xy-bz,$$

## 7.3 Ecuaciones diferenciales de segundo orden.

- Hasta ahora hemos estudiado ecuaciones diferenciales de primer orden. Sin embargo, muchos problemas físicos involucran ecuaciones difrenciables dinámicas de segundo orden (aparecen derivadas segundas) o de orden superior.

- La forma general de una ecuación de segundo orden en una variable es


$$\frac{\text{d}^2x}{\text{d}t^2}=f\left(x,\frac{\text{d}x}{\text{d}t},t\right),$$


$\quad$ es decir una ecuación que puede involucrar, $x$, su derivada y el tiempo.

- La forma de lidiar con estas ecuaciones, es sin embargo, totalmente equivalente a los sistemas de ecuaciones ordiarias cuando uno tiene en cuenta que llamando

$$\frac{\text{d}x}{\text{d}t}=y,$$

$\quad$ nuestra ecuación se puede escribir como

$$\frac{\text{d}y}{\text{d}t}=f(x,y,t),$$

$\quad$ y por tanto, lo convierte en un sistema de dos ecuaciones con dos incognitas.


- De forma analoga, para ecuaciones de orden superior tenemos de forma general el problema

$$\frac{\text{d}^nx}{\text{d}t^n}=f\left(x,\frac{\text{d}x}{\text{d}t},\frac{\text{d}^2x}{\text{d}t^2},\cdots,\frac{\text{d}^{n-1}x}{\text{d}t^{n-1}},t\right),$$

$\quad$ que podemos expresar como

\begin{align}
\frac{\text{d}x}{\text{d}t}=&y_1,\nonumber\\
\frac{\text{d}y_1}{\text{d}t}=&y_2,\nonumber\\
\frac{\text{d}y_2}{\text{d}t}=&y_3,\nonumber\\
\vdots&\nonumber\\
\frac{\text{d}y_{n}}{\text{d}t}=&f(x,y_1,y_2,\cdots,y_{n-1},t).\nonumber\\
\end{align}

- De la misma forma, podemos generalizar un sistema de ecuaciones de orden 2 o superior tratando nuestras variables de forma vectorial; dado el sistema:

$$\frac{\text{d}^2{\bf r}}{\text{d}t^2}=f\left({\bf r},\frac{\text{d}{\bf r}}{\text{d}t},t\right),$$

$\quad$ podemos escribirlo como:

$$\frac{\text{d}{\bf r}}{\text{d}t}={\bf s},\quad \frac{\text{d}{\bf s}}{\text{d}t}=f\left({\bf r},{\bf s},t\right).$$

- De forma general, si comenzamos con un sistema de n ecuaciones de orden m podremos convertirlas en un sistema de $m\times n$ ecuaciones de primer orden, que podremos resolver con los metodos que hemos estudiado hasta ahora.

```python
from numpy import sin,cos,array,linspace,pi
from matplotlib.pyplot import plot,show,xlabel,ylabel

# constantes del sistema

g=9.81
l=0.1

# definimos nuestro intervalo y el paso

a=0
b=10
N=1000
h=(b-a)/N

# definimos nuestra función vectorial

def f(r):
    theta=r[0]
    omega=r[1]
    ftheta=omega
    fomega=-g/l*sin(theta)
    
    return array([ftheta,fomega],float)

# inicializamos nuestras arrays

tp=linspace(a,b,N)
thetap=[]
omegap=[]
xp=[]         # también incluimos la posicion en cartesianas de nuestro péndulo para la animación.
yp=[]

# fijamos las condiciones iniciales

r=array([pi*179/180,0],float)
x=l*sin(r[0])
y=-l*cos(r[0])

for t in tp:
        
    thetap.append(r[0])
    omegap.append(r[1])
    xp.append(x)
    yp.append(y)
    
    k1=h*f(r)
    k2=h*f(r+0.5*k1)
    k3=h*f(r+0.5*k2)
    k4=h*f(r+k3)
    
    r+=(k1+2*k2+2*k3+k4)/6
    
    x=l*sin(r[0])
    y=-l*cos(r[0])
    
plot(tp,thetap,"red")
plot(tp,thetap,"b.")
xlabel("t")
ylabel("$theta$")
show()
```
