## 5.3 Ecuaciones no lineales.

- Por ahora sólo hemos aprendido a resolver sistemas de ecuaciones lineales.
- Pero muchas ecuaciones que aparecen en Física son no-lineales.
- Las ecuaciones no lineales son mucho más complicadas de resolver.
- En particular, resolver una única ecuación no lineal puede ser ya complicado, por lo que resolver varias a la vez aún más díficil.
- En este apartado veremos algunas técnicas para enfrentarnos a estos problemas.
- Empezaremos por el caso de una única ecuación no-lineal.

### 5.3.1 Método del punto fijo o relajación.

- Se aplica a ecuaciones del tipo:
 
$$x=f(x),$$
 
$\quad$ y consiste en busca una solución iterativamente. Es decir, partiendo de una estimación de la soluciónm $x_0$ la siguiente es:

$$x_1=f(x_0),$$

$\quad$ de forma que tras $k$ iteracciones, la solución exacta se ha encontrado si:

$$x_{k+1}-x_k=f\left(x_k\right)-x_k=0.$$

- Lo relevante es por tanto estudiar bajo que condiciones ese proceso iterativo converge a la solución exacta.

- Para ello supongamos que tenemos una estimación de la solución $x_i$ y definamos la solución exacta como $\bar x$. 
-Expandiendo en serie de Taylor en torno a la solución, se tiene:

$$x_{i+1}=f\left(x_i\right)=f\left(\bar x\right)+\left(x_i-\bar x\right)\,f'\left(\bar x\right)+\cdots=\bar x+\left(x_i-\bar x\right)\,f'\left(\bar x\right)+\cdots,$$

$\quad$ por tanto, la diferencia entre nuestra siguiente estimación y la solución exacta es:

$$x_{i+1}-\bar x =(x_i-\bar x)\,f'\left(\bar x\right)+\cdots,$$

- La distancia de las diferentes estimaciones a la solución exacta irá disminyendo si y sólo si

$$\vert f'\left(\bar x\right)\vert<1,$$

**Ejercicio 5.15:** aplicar el método del punto fijo para encontrar la solución a la ecuación

$$x=2-\text{e}^{-x},$$


que no tiene solución analítica.

¿Por qué converge el método?

```python
from numpy import exp

def f(x):
    return 2-exp(-x)

eps=1e-6 # definimos la
it=0
err=1.0
x0=1.0
while err>eps:
    x1=f(x0)
    err=abs(x1-x0)
    it+=1
    x0=x1
    print(it,x1)
```

```python
# Veamos cual es el valor de la derivada en la solución
h=1e-1
derx0=(f(x0+h/2)-f(x0-h/2))/h

print(derx0,exp(-x0),sep=", ")
```

**No convergencia del método del punto fijo**

* Si la derivada de la función en la solución cumple

$$\vert f'(\bar x)\vert>1,$$

$\quad$ el método del punto fijo no convergerá hacía la solución.

* Sin embargo, si la inversa de $f$ existe,

$$x=f(x)\,\Rightarrow\,x=f^{-1}(x),$$

$\quad$ y desarrollando en serie de Taylor en torno a la solución se obtiene:

$$x_{i+1}=f^{-1}\left(x_i\right)=f^{-1}\left(\bar x\right)+(x_i-\bar x)\,f^{-1\,'}\left(\bar x\right)+\cdots=\bar x+(x_i-\bar x)\,\frac{1}{f'(\bar x)}+\cdots,$$

$\quad$ que por tanto convergerá a la solución si:

$$\vert f'\left(\bar x\right)\vert>1.$$

**Ejercicio 5.15:** aplicar el método del punto fijo para encontrar la solución a la ecuación

$$x=\text{e}^{1-x^2}.$$

La ecuación no tiene solución analítica. Sin embargo se ve trivialmente que $x=1$ es una solución.

La derivada de la función en ese punto es:

$$\left.f'(x)\right\vert_{x=1}=\left.-2\,x\,\text{e}^{1-x^2}\;\right\vert_{x=1}=-2,$$

y por tanto, la aplicación directa del método del punto fijo no va a funcionar. 
Comprobemoslo con las 10 primeras itearcciones:

```python
from math import exp
x = 0
eps = 1e-6
for k in range(10):
    x = exp(1-x**2)
    err= abs(x-exp(1-x**2))
    if err < eps:
        break
    print(k+1,x,sep=", ")
```

- La estimación va oscilando continuamente sin obtener ningún patrón de convergencia.

- Haciendo uso de que la inversa de la función existe, podemos reescribir el problema como:

$$x=\sqrt{1-\log x},$$

$\quad$ cuya solución si podemos obtener por el método del punto fijo.  

```python
from numpy import sqrt,log

def f(x):
    return sqrt(1-log(x))

eps=1e-6 # definimos la
it=0
err=1.0
x0=0.5
while err>eps:
    x1=f(x0)
    err=abs(x1-x0)
    it+=1
    x0=x1
    print(it,x1)
```

**Limitaciones del método del punto fijo:**

- Puede haber más de una solución:
<br/>

    - En ese caso el valor de la solución encontrada por el método de relajación dependerá del valor inicial.
    - Así, si tenemos una idea aproximada de la solución que estabamos buscando deberemos usarla para el valor inicial.


- Hay funciones cuya inversa no existe o no sé puede determinar analíticamente. 
<br/>

    - En ese caso, si $|f'(\bar x)|>1$ no será posible encontrar una solución. 
    
    
    
- Sin embargo, hay ocasiones en la que, incluso en ese caso, aún es posible encontrar una solución reordenando la parcialmente la función.

**Ejercicio 5.16:** dada la ecuación:

$$x=x^2+\sin 2x,$$

cuya solución es $x=0$, explicar si es resoluble por medio del método de relajación/punto fijo.   
En caso negativo, intentar reescribir la ecuación de forma que el método converga. 

La derivada de la función en x=0 es $|f'(x)|=2$, por lo que el método de reljación no funcionará.  
Además, la función $f(x)=x^2+\sin 2x$ no es invertible.  
Sin embargo, reescribiendola como:

$$x=g(x)=\frac{1}{2}\sin^{-1}\left(x-x^2\right),$$

se puede comprobar que en ese caso: $|g'(x)|=1/2$ por lo el método del punto fijo en este caso si va a congerger hacía la solución.

```python
from numpy import arcsin

def f(x):
    return arcsin(x-x*x)/2

eps=1e-6 # definimos la
it=0
err=1.0
x0=1e-4
while err>eps:
    x1=f(x0)
    err=abs(x1-x0)
    it+=1
    x0=x1
    print(it,x1)
```

#### Convergencia y estimación del error del método del punto fijo/relajación.

- Una vez hemos visto las condiciones por las cuales el método del punto fijo converge, el paso siguiente es saber como de rápido lo hace.

- Dada una iteracción $i$ la diferencia con la solución exacta es

\begin{align}
|x_{i+1}-\bar x| =|x_i-\bar x\,|\,|f'\left(\bar x\right)|+\cdots=&\;|x_{i-1}-\bar x\,|\,|f'\left(\bar x\right)|^2+\cdots=\cdots=|x_{0}-\bar x\,|\,|f'\left(\bar x\right)\,|^{i+1}+\cdots,\nonumber\\
=\;&|x_0-\bar x\,|\;\text{e}^{(i+1)\log |f'(\bar x)|}+\cdots=|x_0-\bar x\,|\;\text{e}^{-(i+1)\log\left(1/|f'(\bar  x)|\right)}+\cdots
\end{align}

$\quad$ que implica que el método del punto fijo converge exponencialmente.

- Aunque sabemos que el método converge rapidamente, también es interesante saber cuantas iteracciones necesitanamos hasta llegar a una determinada precisión. O lo que es lo mismo, cual es el error de una estimación.


- Dada la estimación $i$-esisma, sabemos que la solución exacta es:

$$\bar x=x_{i}+\epsilon_i=x_{i+1}+\epsilon_{i+1},$$

$\quad$ sin embargo ya hemos visto que: 

$$\epsilon_{i}=\frac{\epsilon_{i+1}}{f'(\bar x)},$$

$\quad$ por lo que se obtiene que

$$x_{i+1}+\epsilon_{i+1}=x_{i}+\frac{\epsilon_{i+1}}{f'(\bar x)}\;\Rightarrow\;\epsilon_{i+1}=\frac{x_{i}-x_{i+1}}{1-1/f'(\bar x)},$$

- Sin embargo, en general no conocemos la solución exacta, que es la que estamos buscando. 

- Aproximando que nuestra estimación $k$-esima no esta lejos de la solución, podemos desarrolar $f'(\bar x)$ en serie de Taylor

$$f'(\bar x) =f'(x_{i})-(\bar x-x_{i})f''(x_{i})+\cdots\simeq f'(x_{i}),$$

$\quad$ que nos permite aproximar que:

$$\epsilon_{i+1}\simeq\frac{x_{i}-x_{i+1}}{1-1/f'(x_{i})}.$$

- Por tanto sólo tenemos que repetir la iteracción hasta que $\epsilon_{i+1}<\epsilon$ siendo $\epsilon$ nuestra tolerancia, sin necesidad de calcular la siguiente iteracción $i+2$.


- Sin embargo, en el caso de no conocer la forma funcional de $f(x)$ tendremos que estimar su derivada numericamente:

$$f'(x_{i})\simeq\frac{f(x_{i})-f(x_{i-1})}{x_{i}-x_{i-1}}=\frac{x_{i+1}-x_{i}}{x_{i}-x_{i-1}},$$

$\quad$ que nos permite aproximar el error de estimación $i+1$

$$\epsilon_{i+1}\simeq\frac{\left(x_{i}-x_{i+1}\right)^2}{2x_i-x_{i+1}-x_{i-1}},$$

$\quad$ a partir de las dos estimaciones anteriores.

**Ejercicio 5.19: El método de sobrerelajación.**

Al igual que con el método de Gauss-seidel para los sistemas de ecuaciones lineales, el método de la sobrerelajación puede ayudar a hacer el método del punto fijo más rápido. 

El método de sobrerelajación acelera la convergencia del sistema iterativo utilizando en cada paso la dirección en que la iteracción esta yendo. Dada la estimación

$$x_{i+1}=f(x_i)$$

la diferencia entre dos estimaciones es

$$\Delta x_{i+1}=x_{i+1}-x_{i}=f(x_i)-x_i,\;\Rightarrow\;x_{i+1}=x_i +\Delta x_{i+1}.$$

El método de sobrerelajación modifica esta ecuación  introduciendo un parámetro $\omega$ de forma que:

$$x_{i+1}=x_i+(1+\omega)\,\Delta x_{i+1}=x_i+(1+\omega)\,(\,f(x_{i})-x_i)=(1+\omega)f(x_i)-\omega\,x_i.$$

Si el parámetro $\omega > 0$ la iteracción cambia por un factor $\omega \Delta x_{i+1}$  más de lo que hace el método de relajación normal.   
Para que el método funcione $\omega$ tiene que escogerse correctamente.

### 5.3.2 Método del punto fijo o relajación con más de una variable.

- El metodo del punto fijo o relajación se puede extender facilmente para el estudio de sistemas de ecuaciones no lineales en varias variables.
- Dadas $N$ ecuaciones con $N$ incongnitas:

\begin{align}
x_1=&f_1(x_1,\cdots,x_N),\nonumber\\
x_2=&f_2(x_1,\cdots,x_N),\nonumber\\
\vdots\quad&\quad\quad\vdots\nonumber\\
x_N=&f_N(x_1,\cdots,x_N),\nonumber\\
\end{align}

$\quad$ partiendo de una estimación inicial para cada una de las variables, $x_i^{(0)}$ con $i=1,\cdots,N$, aplicamos las ecuaciones iterativamente hasta encontrar una solución.

- Como en una dimensión, la convergencia no esta garantizada. 
- En este caso se puede demostar que el sistema convergerá si todos los autovalores $\lambda_i$ de la matriz jacobiana $J$, aquella cuyos elementos de matriz son:

$$J_{k l}=\frac{\text{d}f_k(x_1,\cdots,x_N)}{\text{d}x_l},$$

$\quad$ evaluados en la solución cumplen que $|\lambda_i|<1$.

- Si el sistema no converge, uno puede modificar el sistema de ecuaciones para intentar conseguir que lo haga.

### 5.3.3 Método de la bisección.

- El método de relajación o de punto fijo converge rápidamente y es fácil de programar.

- Sin embargo, sólo funciona en ciertas ocasiones. 

- El método de la bisección, es más robusto para resolver ecuaciones no lineales de una variable. 

- Requiere especificar el intervalo en el que se quiere buscar la solución de una ecuación y  
  garantiza que la encontará siempre y cuando esta sea única. 
  
- Consiste en encontrar una solución de la ecuación:

$$f(x)=0,$$

$\quad$ el método se bada por tanto en encontrar las raíces de una ecuación.

#### Algoritmo para el método de la bisección.

1. Definimos el intervalo en el que queremos encontrar una solución ($x_1$, $x_2$).   
   Comprobamos que los valores de la función en esos puntos, $f(x_1)$ y $f(x_2)$ tienen signos diferentes.  
   En ese caso, sí la función es continua, debe haber como mínimo un punto en el que función se anula (Teorema de Bolzano).  
   Se dice por tanto que el intervalo $[x_1,x_2]$ contiene la raíz. 
   
   
2. Calculamos el punto intermedio del intervalo, $x'=\frac{1}{2}\left(x_1+x_2\right)$, y evaluamos la función $f(x')$.


3. Si $f(x')$ tiene el mismo signo de $f(x_1)$, hacemos la sutición $x_1=x'$. En caso contrario hacemos $x_2=x'$.


4. Si $\vert x_1-x_2\vert>\epsilon$, con $\epsilon$ la precisión objetivo de nuestra solución, volvemos al paso 2.  
   En caso contrario nuestra estimácion de la solución es el nuevo punto medio:   $x'=\frac{1}{2}\left(x_1+x_2\right)$.

  

```python
from IPython.display import display,Image

display(Image(filename="bisection.png", width=600, height=600))
```

#### Convergencia del método de la bisección.

- Al igual que con el método del punto fijo la precisión mejora exponencialmente. 

- La distancia entre dos estimaciones se divide por dos en cada iteración.  
  Por tanto, llamando $\epsilon_0=| x_1-x_2 |$ al error asociado a la estimación inicial, al cabo de $n$ iteraciones:
  
  
  $$\epsilon_n=\frac{\epsilon_0}{2^n}=\epsilon_0 \text{e}^{-n\log2}.$$
  
  
- Esto nos permite calcular el número de iteracciones necesarias $N$ para llegar a una precisión obetivo $\epsilon$:

$$N=\log_2\frac{\epsilon_0}{\epsilon}.$$

- Como el logaritmo es una función que crece muy lentamente, esto nos asegura que con un número de pasos razonable podremos encontrar la solución con una determinada precisión. 

- Por ejemplo, sí $\epsilon_0=\vert x_1-x_2\vert = 10^{10}$ y nuesra precisión objetivo $\epsilon=10^{-10}$ el número de iteracciones necesarias será:

```python
from math import log 

log(10**20,2)
```

#### Limitaciones del método de la bisección.

- Si dado nuestro intervalo inicial $f(x_1)$ y $f(x_2)$ tienen el mismo signo el método no funciona. 


- Además esto se puede deber a que no hay solución, o que hay un número par de soluciones.   
  Lamentablemebte no es fácil distinguir entre ambas posibilidades. 

```python
display(Image(filename="bisection2.png", width=600, height=600))
```

- Igualmente, es muy posible que no se tenga una estimación inicial del intervalo en el que debe encontrarse la raíz.     
   Cualquier par de puntos para los que la función tenga diferente signo valdrán, pero encontrarlos puede no ser fácil.
   
 
- Reglas generales que facilitan encontrar el intervalo inicial son:
<br/>

    - Si se tiene información sobre la función que permite encontrar donde puede estar la raíz, tenerla en cuenta.  
      Por ejemplo,  si se conoce el signo de la función por consideraciones físicas en un determinado intervalo.
<br/>

    - Si se busca una solución en torno a un punto eso ayuda.   
      Tomamos entonces un entorno alrededor del punto y vamos aumentando el tamaño hasta que contenga una raíz.


- De la misma forma, el método no permite encontrar soluciones de ráices múltiples pares.  
  Estos problemas son equivalentes a encontrar el mínimo de una función.  

```python
display(Image(filename="bisection3.png", width=600, height=600))
```

- Finalmente, es un método que no se puede generalizar a funciones de más de una variable. 

**Ejercicio 5.21: Visualizando el método de la bisección.** 

Encontrar la raíz de la función:

$$f(x)=\text{e}^x-2,$$

usando el método de la biscección con una precisión mayor de $10^{-6}$.

```python
from numpy import exp,linspace
from matplotlib.pyplot import plot,show,axhline,subplots

def f(x):
    return exp(x)-2

x=linspace(-2,2,100)
fig,ax=subplots(1,1,figsize=(12, 4)) # subplots(#filas,#columas) 
                                     # nos permite crear varias gráficas en una sólo llamada.
                                     # y cambiar facilmente sus propiedades    
ax.plot(x,f(x),lw=1.5)
ax.axhline(0,ls=':',color="k")
ax.set_xticks([-2, -1, 0, 1, 2])
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$f(x) = e^x - 2$', fontsize=18)
show()
```

```python
# En el intervalo [-2,2] tenemos seguro como mínimo una raíz.

# Definimos nuestro intervalo inicial
a,b=-2,2
fa,fb=f(a),f(b)

fig2,ax2=subplots(1,1,figsize=(12, 4))     

ax2.plot(x,f(x),lw=1.5)

# Lo añadimos a la figura.

ax2.plot(a,fa, 'ko')
ax2.plot(b,fb, 'ko')
ax2.text(a,fa + 0.5, r"$a$", ha='center', fontsize=18)
ax2.text(b,fb + 0.5, r"$b$", ha='center', fontsize=18)
show()
```

```python
from numpy import sign

# Definimos nuestra tolerancia
eps=1e-6

# Encontramos la raíz usando el método de la bisección y visualizamos
# los pasos del método en el gráfico

fig3,ax3=subplots(1,1,figsize=(12, 4)) 
ax3.plot(x,f(x),lw=1.5)
ax3.plot(a,fa, 'ko')
ax3.plot(b,fb, 'ko')
ax3.text(a,fa + 0.5, r"$a$", ha='center', fontsize=18)
ax3.text(b,fb + 0.5, r"$b$", ha='center', fontsize=18)

it=0

while b-a>eps:
    m=a+(b-a)/2
    fm=f(m)
    ax3.plot(m,fm,'ko')
    ax3.text(m,fm-0.5, r"$m_%d$" %it,ha='center')

    it += 1    
    if sign(fa)==sign(fm):
        a,fa=m,fm
    else:
        b,fb= m,fm
print(it)
        
ax3.plot(m,fm,'r*',markersize=10)

ax3.annotate("Raíz aproximada en %.3f" % m,fontsize=14, family="serif",
xy=(a, fm), xycoords='data',
xytext=(-150, +50), textcoords='offset points',
arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=-.5"))

ax3.set_title("Método de la Bisección")

show()
```
